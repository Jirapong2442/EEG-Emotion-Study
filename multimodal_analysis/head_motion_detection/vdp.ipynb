```python
import numpy as np
import pandas as pd
import os
```

```python
path_to_folder = "C:/Users/NA/Saved Games/short_vdo/EMO-AffectNetModel-main/report"
files = os.listdir(path_to_folder)
data = {}
for file in files:
    file_path = os.path.join(path_to_folder,file)
    data[file] = pd.read_csv(file_path)
```

```python
files
```

```python
import matplotlib.pyplot as plt
x_col = data['ice_test.csv'].columns[0]
y_cols = data['ice_test.csv'].columns[1:]
plt.figure(figsize=(10, 6))
for col in y_cols:
    plt.plot(data['ice_test.csv'][x_col], data['ice_test.csv'][col], label=col, marker='o', markersize = 1, linewidth = 1)


plt.xlabel(x_col, fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.title('Probability of facial expression class', fontsize=14)
plt.legend(title='Columns')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to prevent label cutoff
plt.show()
```
